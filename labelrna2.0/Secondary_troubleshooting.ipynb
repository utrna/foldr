```python
import pandas as pd
import numpy as np
alden = pd.read_csv("../piesie2/Ecoli_16s-rRNA.labels.csv", header=None)
alden.insert(3, "ie", np.nan)
alden.insert(3, "ps", np.nan)
alden.insert(3, "label", np.nan)
alden[:10]
```

```python
hairpins = alden[alden[2] == "HAIRPIN"]
i = 1
for row in hairpins.itertuples():
    j = row[8]
    alden.ix[alden[3]==j, "ps"] = "PRIMARY"
    alden.ix[alden[3]==j, "ie"] = "INITIATION"
    alden.ix[alden[3]==j, "label"] = 'P'+str(i)+'a'
    i += 1
alden[:10]
```

```python
for row in hairpins.itertuples():
    i = row[0] + 1
    j = row[0] - 1
    while i < len(alden.index) and alden.ix[i, 2] != "MULTISTEM" and alden.ix[i, 2] != "FREE" and j > 0 and alden.ix[j, 2] != "MULTISTEM" and alden.ix[j, 2] != "FREE":
        if alden.ix[i, 2] == "HELIX" and alden.ix[i, "ie"] != "INITIATION":
            l = alden.ix[i, 3]
            alden.ix[alden[3]==l, "ps"] = "PRIMARY"
            alden.ix[alden[3]==l, "ie"] = "ELONGATION"
        i += 1
        j -= 1
alden[:10]
```

```python
alden1 = pd.read_csv("../piesie2/prim_16_correct.alps", sep=' ', header=None)
is_helix = alden1[2] == "HELIX"
not_assigned = alden1[4] != "PRIMARY"
sec = alden1[is_helix & not_assigned]
sec[:10]
```

```python
from string import ascii_lowercase
stack = []
i = 1
j = 0
for row in sec.itertuples():
    if len(stack) > 0 and row[7] == stack[len(stack)-1][7]:
        alden1.ix[alden1[6]==row[7], 4] = "SECONDARY"
        if alden1.ix[row[0]+1, 2] != "MULTISTEM" and alden1.ix[row[0]+1, 2] != "FREE" and alden1.ix[stack[len(stack)-1][0]-1, 2] != "MULTISTEM" and alden1.ix[stack[len(stack)-1][0]-1, 2] != "FREE":
            if j == 0:
                alden1.ix[alden1[6]==row[7], 5] = "INITIATION"
                alden1.ix[alden1[6]==row[7], 3] = 'S'+str(i)+ascii_lowercase[j]
                j += 1
            else:
                alden1.ix[alden1[6]==row[7], 5] = "ELONGATION"
                alden1.ix[alden1[6]==row[7], 3] = 'S'+str(i)+ascii_lowercase[j]
                j += 1
        else:
            if j == 0:
                alden1.ix[alden1[6]==row[7], 5] = "INITIATION"
                alden1.ix[alden1[6]==row[7], 3] = 'S'+str(i)+ascii_lowercase[j]
                j = 0
            else:
                alden1.ix[alden1[6]==row[7], 5] = "ELONGATION"
                alden1.ix[alden1[6]==row[7], 3] = 'S'+str(i)+ascii_lowercase[j]
                j = 0
            i += 1
        stack.pop()
    else:
        stack.append(row)
alden1[alden1[4] == "SECONDARY"][:20]
```
